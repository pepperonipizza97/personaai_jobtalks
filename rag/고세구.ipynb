```python
from vllm import LLM, SamplingParams


llm = LLM(
    model="CohereForAI/c4ai-command-r-plus",
    tensor_parallel_size=8,
    max_model_len=4096,
    gpu_memory_utilization=0.5,
    dtype="float16",
)
```

```python
# 텍스트 생성을 위한 매개변수 설정
params = SamplingParams(
    max_tokens=100,            # 생성할 최대 토큰 수
    temperature=0.7,           # 샘플링 온도, 낮을수록 예측 가능한 텍스트 생성
    top_p=0.9                  # top-p 샘플링 확률
)

# 텍스트 생성 요청
prompt_text = "오늘의 날씨는 어떤가요?"
generated_text = llm.generate(prompt_text, params)

# 생성된 텍스트 출력
print(generated_text)
```

```python
from llama_index.llms.vllm import Vllm
import torch
from multiprocessing import set_start_method


torch.multiprocessing.set_start_method('spawn')


llm = Vllm(
    model="CohereForAI/c4ai-command-r-plus",
    tensor_parallel_size=8,
    max_new_tokens=100,
    vllm_kwargs={"swap_space": 1, "gpu_memory_utilization": 0.8},
)

```

```python
llm.complete(
    "what is banana? answer is korean"
)
```
